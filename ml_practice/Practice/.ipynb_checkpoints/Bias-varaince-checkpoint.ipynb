```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
df= pd.read_csv("Advertising.csv")
```

```python
df.columns
```

```python
df.drop("Unnamed: 0", axis = 1 , inplace = True)
```

```python
X= df.drop("sales", axis = 1)
y= df['sales']
```

```python
test_rmse=[]
train_rmse=[]

for d in range(1,10):
    poly_converter =PolynomialFeatures(degree = d, include_bias= False)
    poly_fit = poly_converter.fit(X)
    poly_features= poly_fit.transform(X)

    #train and test split
    X_train,X_test,y_train,y_test= train_test_split(poly_features,y,test_size=0.3,random_state=101)
    model = LinearRegression()
    model.fit(X_train,y_train)

    test_predication = model.predict(X_test)
    train_predication = model.predict(X_train)

    #Finding the root mean squared of both train and test predication]
    rmse_train = np.sqrt(mean_squared_error(y_train,train_predication))
    rmse_test= np.sqrt(mean_squared_error(y_test,test_predication))

    #listing all the  rmse data in list
    test_rmse.append(rmse_test)
    train_rmse.append(rmse_train)


    
```

```python
test_rmse
```

```python
train_rmse
```

```python
#plotting the model complexity vs error
plt.plot(range(1,6),test_rmse[:5], label='test_rmse')
plt.plot(range(1,6),train_rmse[:5], color='red', label='train_rmse')
plt.legend()
plt.xlabel("Model_complexity")
plt.ylabel("RMSE error")
plt.grid()
plt.show()
```

```python
poly_converter = PolynomialFeatures(degree=3, include_bias=False)
poly_features = poly_converter.fit_transform(X)
```

```python
poly_features.shape
```

```python
#Train and Test Spilt
X_train,X_test,y_train,y_test= train_test_split(poly_features,y,test_size=0.3,random_state=101)
```

```python
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
test_predication = model.predict(X_test)
```

```python
test_residual = y_test-test_predication
```

```python
sns.scatterplot(x=test_predication, y=test_residual)
plt.axhline(y=0, color='red', ls ='--')
```

```python
sns.displot(test_residual,kind='kde')
plt.grid()
plt.show()
```

# Conclusion
- <b>All the plots: `model complexity vs error`, `residual plot (test_prediction vs residual)` and `distribution plot` concludes the model selection is correct with polynomial degree 3.<b>
