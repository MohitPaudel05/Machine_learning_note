```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python

x = np.array([1,2])
y =  np.array([300,500])
# y = wx+b
w=0
b=0
learning_rate=0.01
iterations= 1000
# Gradiant Descent
for i in range(iterations):
    y_pred= w*x+b
    # compute the gradients
    dw = (1/len(x)) * np.sum(x * (y_pred-y))  # Gradient for w
    db = (1/len(x)) * np.sum(y_pred-y) 
    w-=learning_rate*dw
    b-=learning_rate*db

print(w,b)
     
```
